```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt

import tensorflow as tf
import models_tf as models
```

```python
data = sio.loadmat("../solution.mat")
c0 = data["IC"]
t = data["t"]
c_ref = data["y"]

t_train = t

scale = np.array(
    [10, 1, 100, 100, 0.001, 10, 0.1, 1000]
).reshape([-1])
c_ref = c_ref / scale
c0 = c0 / scale
```

```python
np.max(data["y"], axis=0)
```

```python
model = models.PINN(
    c0=c0,
    output_dim=8,
    scale=scale,
    units=100, 
    activation=tf.tanh,
    eps=1e-4,
)
```

```python
t_ode = t_train[::10]
t_u = t_train
u = c_ref
loss, data_loss, ode_loss = model.train(t_ode, t_u, u, niter=120000)
```

```python
model.restore()
c_pred = model.call(
    tf.constant(t_train, tf.float32),
)
```

```python
L2 = np.sqrt(np.sum((c_pred - c_ref) ** 2, axis=0) / np.sum(c_ref ** 2, axis=0))
```

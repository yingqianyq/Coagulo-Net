```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
scale = np.array(
    [10, 1, 100, 100, 0.0001, 10, 0.1, 1000]
).reshape([-1])

# data = sio.loadmat("../test.mat")
data = sio.loadmat("../../data/data_2k_points.mat")
t_train = data["t"]
c_original = data["y"] / scale

c_k8_120 = sio.loadmat("../../data/data_k8_120.mat")["y"]/scale
c_k8_80 = sio.loadmat("../../data/data_k8_80.mat")["y"]/scale
c_h5_120 = sio.loadmat("../../data/data_h5_120.mat")["y"]/scale
c_h5_80 = sio.loadmat("../../data/data_h5_80.mat")["y"]/scale
```

```python
sparsity = [0.0, 0.25, 0.5, 0.75]
k8_inf = [9.997982e-06, 9.978739e-06, 8.768697e-06, 8.683059e-06]
ka_inf = [1.2007586, 1.2026414, 1.2120042, 1.167961]
k5_inf = [0.16980526, 0.1707336, 0.17144364, 0.16086532]
h5_inf = [0.30525765, 0.3134318, 0.29580647, 0.27529213]
k8_ref = [0.00001] *4
ka_ref = [1.2] * 4
k5_ref = [0.17] * 4
h5_ref = [0.31] * 4
y_k8 = np.divide(np.abs(np.subtract(k8_inf, k8_ref)), k8_ref)
y_ka = np.divide(np.abs(np.subtract(ka_inf, ka_ref)), ka_ref)
y_k5 = np.divide(np.abs(np.subtract(k5_inf, k5_ref)), k5_ref)
y_h5 = np.divide(np.abs(np.subtract(h5_inf, h5_ref)), h5_ref)
plt.figure(dpi=120)
plt.plot(np.multiply(sparsity, 100), y_k8, ".--", label="k8", linewidth=2)
plt.plot(np.multiply(sparsity, 100), y_ka, ".--", label="ka", linewidth=2)
plt.plot(np.multiply(sparsity, 100), y_k5, ".--", label="k5", linewidth=2)
plt.plot(np.multiply(sparsity, 100), y_h5, ".--", label="h5", linewidth=2)
plt.plot(np.multiply(sparsity, 100), np.average([y_ka, y_k8, y_k5, y_h5], axis=0), ".-", label="average", linewidth=2)
plt.legend(title="parameter:")
plt.xticks([0, 25, 50, 75])
plt.xlabel("sparsity (%)", fontsize=12)
plt.ylabel("relative error", fontsize=12)
# plt.title("Performance of parameter inference with increasing data sparsity")
plt.show()
```

```python
def read_res(ref_value, index_in_res, file_path, num_of_lines=10):
    ref = [ref_value] * num_of_lines
    inf = []
    f = open(file_path + "res_2k.txt", "r")
    results = f.read().split('\n')
    for res in results:
        res = res.split(',')
        inf.append(float(res[index_in_res]))
    f.close()
    err_2k = np.divide(np.abs(np.subtract(inf, ref)), ref)

    inf = []
    f = open(file_path + "res_1500.txt", "r")
    results = f.read().split('\n')
    for res in results:
        res = res.split(',')
        inf.append(float(res[index_in_res]))
    f.close()
    err_1500 = np.divide(np.abs(np.subtract(inf, ref)), ref)

    inf = []
    f = open(file_path + "res_1k.txt", "r")
    results = f.read().split('\n')
    for res in results:
        res = res.split(',')
        inf.append(float(res[index_in_res]))
    f.close()
    err_1k = np.divide(np.abs(np.subtract(inf, ref)), ref)

    inf = []
    f = open(file_path + "res_500.txt", "r")
    results = f.read().split('\n')
    for res in results:
        res = res.split(',')
        inf.append(float(res[index_in_res]))
    f.close()
    err_500 = np.divide(np.abs(np.subtract(inf, ref)), ref)
    return [err_2k, err_1500, err_1k, err_500]
```

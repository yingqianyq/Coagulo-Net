```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error as mae

import torch
import torch.nn as nn
from torch.autograd import grad

import models
```

```python
# data = sio.loadmat("data/data.mat")
# data = sio.loadmat("data/data_7d_smoothed.mat")
data = sio.loadmat("data/data_weekly_avg.mat")

res = data["res"].T
doses = data["doses"].T
Zr = data["cases"].T
Dr = data["deaths"].T
A = data["hospitalized"].T
t = data["time"].T

Zr[Zr <= 0] = 0 # 1e-8
Dr[Dr <= 0] = 0 #1e-8
A[A <= 0] = 0 #1e-8
```

```python
plt.plot(t, Zr, 'o-')
plt.title("Cases")
plt.show()

plt.plot(t, Dr, 'o-')
plt.title("Deaths")
plt.show()

plt.plot(t, A, 'o-')
plt.title("Hospitalized")
plt.show()
```

```python
# first K weeks' data for training
K = 60
res = res[:K]
doses = doses[:K]
Zr_train = Zr[:K]
Dr_train = Dr[:K]
A_train = A[:K]

t_Zr = t[:K]
t_Dr = t[:K]
t_A = t[:K]

t_ode = t[K:K+4]
t_test = t_ode
Zr_test = Zr[K:K+4]
Dr_test = Dr[K:K+4]
A_test = A[K:K+4]
t_A_test = t_test[~np.isnan(A_test)].reshape([-1, 1])

# preprocess nan
doses[np.isnan(doses)] = 0
t_A = t_A[~np.isnan(A_train)].reshape([-1, 1])
A_train = A_train[~np.isnan(A_train)].reshape([-1, 1])
A_test = A_test[~np.isnan(A_test)].reshape([-1, 1])


N = 39512223.0
eta = 1/4
gamma = 1/4
gamma_dw = 1/10
gamma_zw = 1
gamma_h = 1/10
rho = 0.5
```

```python
# prepare data
t_ode = torch.tensor(t_ode, dtype=torch.float32, requires_grad=True)
t0 = torch.tensor([[0]], dtype=torch.float32)
t_Zr = torch.tensor(t_Zr, dtype=torch.float32)
t_Dr = torch.tensor(t_Dr, dtype=torch.float32)
t_A = torch.tensor(t_A, dtype=torch.float32)

Zr0 = torch.tensor([[0]], dtype=torch.float32)
Dr0 = torch.tensor([[0]], dtype=torch.float32)
A0 = torch.tensor([[0]], dtype=torch.float32)
Zr_train = torch.tensor(Zr_train, dtype=torch.float32)
Dr_train = torch.tensor(Dr_train, dtype=torch.float32)
A_train = torch.tensor(A_train, dtype=torch.float32)
```

```python
# prepare model
u = models.NN(
    layers=[1, 50, 50, 50, 50, 50, 9],
    activation=torch.tanh,
    input_transform=lambda t: t/t.shape[0],
    output_transform=torch.exp,
)
L0 = torch.tensor([[1.0]], dtype=torch.float32, requires_grad=True)
log_beta = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
logit_p_h = torch.tensor(0.5, dtype=torch.float32, requires_grad=True)
logit_p_d = torch.tensor(0.5, dtype=torch.float32, requires_grad=True)

parameters = list(u.parameters()) + [L0, log_beta, logit_p_h, logit_p_d]
opt = torch.optim.Adam(parameters, lr=0.001)
```

```python
weights = torch.tensor([1, 1e-10, 1e-2, 1, 1], dtype=torch.float32)
```

```python
loss_total = []
loss1 = []
loss2 = []
loss3 = []
loss4 = []
loss5 = []
min_loss = 1e10
```

```python
for it in range(50000):
    beta = torch.clip(torch.exp(log_beta), max=1000)
    p_h = torch.sigmoid(logit_p_h)
    p_d = torch.sigmoid(logit_p_d)
    args = [
        beta,
        p_h,
        N, eta, gamma, gamma_dw, gamma_zw, gamma_h,
        p_d,
        rho,
    ]
    Y0 = eta * L0 / gamma
    Z0 = (p_h + rho * (1 - p_h)) * gamma * Y0 / gamma_zw
    H0 = p_h * gamma * Y0 / gamma_h
    D0 = p_d * gamma_h * H0 / gamma_dw
    X0 = torch.clip(N - L0 - Y0 - H0 - D0, min=N/10)
    u0 = torch.concat([X0, L0, Y0, Z0, Zr0, H0, A0, D0, Dr0], dim=-1)
    
    losses = models.loss_function(
        u, 
        t_ode, 
        t0,
        u0,
        t_Zr, Zr_train, 
        t_Dr, Dr_train, 
        t_A, A_train, 
        args,
    )
    loss_value = torch.sum(weights * torch.stack(losses))
    opt.zero_grad()
    loss_value.backward()
    opt.step()
    loss_total += [loss_value.detach().cpu().numpy()]
    # loss_ode, loss_init, loss_Zr, loss_Dr, loss_A
    loss1 += [losses[0].detach().cpu().numpy()]
    loss2 += [losses[1].detach().cpu().numpy()]
    loss3 += [losses[2].detach().cpu().numpy()]
    loss4 += [losses[3].detach().cpu().numpy()]
    loss5 += [losses[4].detach().cpu().numpy()]
    if it % 1000 == 0:
        print(it, 
              loss_total[-1],
              loss1[-1], 
              loss2[-1], 
              loss3[-1], 
              loss4[-1], 
              loss5[-1]
             )
        if loss_total[-1] < min_loss:
            min_loss = loss_total[-1]
            torch.save(u.state_dict(), './checkpoints/model_weights')
```

```python
u.load_state_dict(torch.load('./checkpoints/model_weights'))
u_pred = u.forward(t_ode).detach().cpu()
```

```python
plt.figure(dpi=100)
plt.plot(t_ode.detach().cpu(), u_pred[:, 4:5], '--', label="prediction")
plt.plot(t_Zr, Zr_train, 'o', markersize=1, label="observations")
plt.plot(t_ode.detach().cpu(), Zr_test, label="reference")
plt.title("Cases")
plt.legend()
```

```python
mae_cases = mae(Zr_test, u_pred[:, 4:5])
print("mae of cases: ", mae_cases)

mae_cases_naive_1 = mae(Zr_test, Zr_train[-4:])
print("mae of naive cases 1: ", mae_cases_naive_1)

mase_cases_1 = mae_cases / mae_cases_naive_1
print("mase of cases 1: ", mase_cases_1)
```

```python
plt.figure(dpi=100)
plt.plot(t_ode.detach().cpu(), u_pred[:, 8:9], '--', label="prediction")
plt.plot(t_Dr, Dr_train, 'o', markersize=1, label="observations")
plt.plot(t_ode.detach().cpu(), Dr_test, label="reference")
plt.title("Deaths")
plt.legend()
```

```python
mae_deaths = mae(Dr_test, u_pred[:, 8:9])
print("mae of deaths: ", mae_deaths)

mae_deaths_naive_1 = mae(Dr_test, Dr_train[-4:])
print("mae of naive deaths 1: ", mae_deaths_naive_1)

mase_deaths_1 = mae_deaths / mae_deaths_naive_1
print("mase of deaths 1: ", mase_deaths_1)
```

```python
plt.figure(dpi=100)
plt.plot(t_ode.detach().cpu(), u_pred[:, 6:7], '--', label="prediction")
plt.plot(t_A, A_train, 'o', markersize=1, label="observations")
plt.plot(t_A_test, A_test, label="reference")
plt.title("Hospital Admissions")
plt.legend()
plt.show()
```

```python
mae_hospitalized = mae(A_test, u_pred[:, 6:7])
print("mae of hospitalized: ", mae_hospitalized)

mae_hospitalized_naive_1 = mae(A_test, A_train[-4:])
print("mae of naive hospitalized 1: ", mae_hospitalized_naive_1)

mase_hospitalized_1 = mae_hospitalized / mae_hospitalized_naive_1
print("mase of hospitalized 1: ", mase_hospitalized_1)
```

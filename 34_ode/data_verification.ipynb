```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
data_1 = sio.loadmat("./data/data_case_1.mat")
t_1 = data_1["t"]
u_1 = data_1["Y"]
data_2 = sio.loadmat("./data/data_case_2.mat")
t_2 = data_2["t"]
u_2 = data_2["Y"]
data_3 = sio.loadmat("./data/data_case_3.mat")
t_3 = data_3["t"]
u_3 = data_3["Y"]
data_4 = sio.loadmat("./data/data_case_4.mat")
t_4 = data_4["t"]
u_4 = data_4["Y"]
data_5 = sio.loadmat("./data/data_case_5.mat")
t_5 = data_5["t"]
u_5 = data_5["Y"]
data_6 = sio.loadmat("./data/data_case_6.mat")
t_6 = data_6["t"]
u_6 = data_6["Y"]

#name_list = ['TF','VII', 'TF:VII', 'VIIa', 'TF:VIIa', 'IX', 'IXa', 'IXm', 'IXam',
#'X', 'Xa', 'Xm', 'Xam', 'II', 'IIa', 'IIm', 'IIam', 'PL', 'AP', 'VIII', 'VIIIa', 'VIIIm',
#'VIIIam', 'VIIIam:IXam', 'V', 'Va', 'Vm', 'Vam', 'Xam:Vam', 'I', 'Ia', 'TFPI', 'Xa:TFPI', 'ATIII']

TIM = 1800
NDM = [0.025, 10.0, 0.005, 10.0, 0.005, 90.0, 90.0, 10.0, 10.0, 170.0,
       170.0, 10.0, 10.0, 1400.0, 1400.0, 10.0, 10.0, 10.0, 10.0,
       0.7, 0.7, 0.7, 0.7, 0.7, 20.0, 20.0, 10.0, 10.0, 10.0, 7000.0,
       7000.0, 2.5, 2.5, 3400.0] #nM
```

```python
exp_data_1 = np.loadtxt("./data/exp_data_case_1.dat").T
t_exp_1 = exp_data_1[0]
u_exp_1 = exp_data_1[1]
exp_data_2 = np.loadtxt("./data/exp_data_case_2.dat").T
t_exp_2 = exp_data_2[0]
u_exp_2 = exp_data_2[1]
exp_data_3 = np.loadtxt("./data/exp_data_case_3.dat").T
t_exp_3 = exp_data_3[0]
u_exp_3 = exp_data_3[1]
exp_data_4 = np.loadtxt("./data/exp_data_case_4.dat").T
t_exp_4 = exp_data_4[0]
u_exp_4 = exp_data_4[1]
exp_data_5 = np.loadtxt("./data/exp_data_case_5.dat").T
t_exp_5 = exp_data_5[0]
u_exp_5 = exp_data_5[1]
exp_data_6 = np.loadtxt("./data/exp_data_case_6.dat").T
t_exp_6 = exp_data_6[0]
u_exp_6 = exp_data_6[1]
```

```python
plt.plot(t_1 * TIM, (u_1[:, 14] * NDM[14] + u_1[:, 16] * NDM[16]) * 1e-9, 'b', label='Model(No Inh)')
plt.plot(t_exp_1, u_exp_1, 'bo', markerfacecolor='w', label='Exp')
plt.plot(t_2 * TIM, (u_2[:, 14] * NDM[14] + u_2[:, 16] * NDM[16]) * 1e-9, 'r', label='Model(ATIII alone)')
plt.plot(t_exp_2, u_exp_2, 'rD', markerfacecolor='w', label='Exp')
plt.plot(t_3 * TIM, (u_3[:, 14] * NDM[14] + u_3[:, 16] * NDM[16]) * 1e-9, 'k', label='Model(Both Inhs)')
plt.plot(t_exp_3, u_exp_3, 'ks', markerfacecolor='w', label='Exp')
plt.legend()
plt.title("Data generated by myself\n(Model data is generated by matlab and Exp data is generated by imageread)")
plt.show()
```

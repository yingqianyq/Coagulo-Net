```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
data_ori = sio.loadmat("./data/ori_k14.mat")
data_1000 = sio.loadmat("./data/1000_k14.mat")
data_001 = sio.loadmat("./data/001_k14.mat")
data_0 = sio.loadmat("./data/0_k14.mat")
```

```python
plt.figure(dpi=120)
plt.plot(data_ori['save_t'], data_ori['save_thrombin'], label="original $\mathregular{h_{10}^{TP-}}$", alpha = 0.6)
plt.plot(data_1000['save_t'], data_1000['save_thrombin'], '--', label="1000 * $\mathregular{h_{10}^{TP-}}$", alpha = 0.6)
plt.plot(data_001['save_t'], data_001['save_thrombin'], '--', label="0.001 * $\mathregular{h_{10}^{TP-}}$", alpha = 0.6)
plt.plot(data_0['save_t'], data_0['save_thrombin'], '--', label="0 * $\mathregular{h_{10}^{TP-}}$", alpha = 0.6)
plt.xlabel("Time (s)")
plt.ylabel("$\mathregular{IIa+IIa^m}$ (M)")
plt.legend()
plt.show()
```

```python
plt.figure(dpi=120)
plt.plot(data_ori['save_t'], data_ori['save_thrombin'] - data_1000['save_thrombin'], label="original - 1000 * value")
plt.plot(data_ori['save_t'], data_ori['save_thrombin'] - data_001['save_thrombin'], label="original - 0.001 * value")
plt.plot(data_ori['save_t'], data_ori['save_thrombin'] - data_0['save_thrombin'], label="original - 0 * value")
plt.xlabel("Time(s)")
plt.ylabel("thrombin(M)")
plt.legend()
plt.show()
```
